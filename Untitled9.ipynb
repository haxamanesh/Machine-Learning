```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
```

```python
df1 = df.drop(["Revenue"], axis=1)
df1.head()
```

```python
objects = df1.select_dtypes(include=['object']).columns
bools = df1.select_dtypes(include=['bool']).columns
num = df1.select_dtypes(exclude = ['object','bool'] ).columns
print(num)
```

```python
for col in objects:
  print(df1[col].unique())
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num),
        ('bool',"passthrough", bools)]
    )

df2 = preprocessor.fit_transform(df1)
df2 = pd.DataFrame(df2)
df2.head()
```

```python
plt.figure(figsize=(13, 13))
sns.heatmap(df2.corr(), annot=False, cmap='coolwarm')
plt.show()
```

```python
pca = PCA(n_components=17)
kmeans = KMeans(n_clusters=5, random_state=42)  # Example: 5 clusters

pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('pca', pca),
                           ('clusterer', kmeans)])

# Fit the pipeline
pipeline.fit(df1)

# Access cluster labels
cluster_labels = pipeline.named_steps['clusterer'].labels_
```

```python
pca_model = pipeline.named_steps['pca']
pca_components = pipeline.named_steps['pca'].transform(preprocessor.transform(df1))
explained_variance_ratio = pca_model.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

pca_components = pca_model.components_
pca_df1 = pd.DataFrame(data=pca_model.transform(preprocessor.transform(df1.drop('cluster', axis=1))),
                      columns=[f'PC{i+1}' for i in range(pca_model.n_components_)])
```
